```python
import os
import librosa  
import soundfile as sf
import time

# sr = sample rate
# sr=192000
sr=384000
noise_file = '/home/paddy/Desktop/noise_01.wav'

def renameFiles(path, basefilename, depth=1):
    # Once we hit depth, just return (base case)
    if depth < 0: return

    # Make sure that a path was supplied and it is not a symbolic link
    if os.path.isdir(path) and not os.path.islink(path):
        ind = 1

        # Loop through each file in the start directory and create a fullpath
        for file in os.listdir(path):
            fullpath = path + os.path.sep + file

            # Again we don't want to follow symbolic links
            if not os.path.islink(fullpath):

                # If it is a directory, recursively call this function 
                # giving that path and reducing the depth.
                if os.path.isdir(fullpath):
                    renameFiles(fullpath, basefilename, depth - 1)
                else:
                    # Find the extension (if available) and rebuild file name 
                    # using the directory, new base filename, index and the old extension.
                    extension = os.path.splitext(fullpath)[1]
                    os.rename(fullpath, os.path.dirname(fullpath) + os.path.sep + basefilename + "_" + str(ind) + extension)
                    
                    import_file = (os.path.dirname(fullpath) + os.path.sep + basefilename + "_" + str(ind) + extension)
                    print(import_file)

                    data, rate = librosa.load(import_file, sr) # Resample to 384000 Hz
                    data = data
                    noise_clip, rate = librosa.load(noise_file, sr) # Resample to 384000 Hz
                    noise_clip = noise_clip
                    # Remove noise:
                    noise_reduced = nr.reduce_noise(audio_clip=data, noise_clip=noise_clip, verbose=False)
                    export_file = import_file
                    sf.write(export_file, noise_reduced, sr, format='wav', subtype='PCM_16')
                    ind += 1
            
    return

renameFiles('/home/paddy/Desktop/TEST/c_pip/', "c_pip_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/s_pip/', "s_pip_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/ferrum_rhino/', "ferrum_rhino_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/hippo_rhino/', "hippo_rhino_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/nattereri/', "nattereri_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/noctula/', "noctula_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/n_pip/', "n_pip_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/plecotus/', "plecotus_noise_reduced", depth=1)
renameFiles('/home/paddy/Desktop/TEST/rodents/', "rodents_noise_reduced", depth=1)

```
